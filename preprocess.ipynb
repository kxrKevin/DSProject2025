## Preprocessing Rationale:
There are 7 String columns and 2 columns that contain negative values out of 726 numeric columns. Since the proportion of negative values for HSWH041S (net purchase price of secondary owned residences) is very low, the negative entries will be imputed for HSWH041S while the other should be imputed. Among the 2 negative numeric variables representing net home price for secondary and first hand homes, there are an equal number of 0s, so they should be treated as null values and imputed.  

All string columns are columns for median values from Demo Stats, with values stored as strings and not applicable values as "NA". "NAs" will be treated as null values and removed or imputed according to the proportion that are null values.

```python
# check the String columns and process it

string_cols = [col for col, dtype in df.schema.items() if dtype == pl.Utf8]
df_strings = df.select(string_cols)
print(df_strings.columns)

total_rows = len(df)
for col in string_cols: # count the number of NAs
    na_count = df.filter(pl.col(col) == "NA").height
    print(f'"NA" count: {na_count} / {total_rows}')
    print(f"percentage of nulls for {col}: {na_count / total_rows * 100}")
    
# cast the "NA"s to nulls first before casting them to int64
for col in string_cols:
    df = df.with_columns(
    pl.when(pl.col(col) == "NA")
      .then(None)
      .otherwise(pl.col(col))
      .alias(col)
    )

    df = df.with_columns(pl.col(col).cast(pl.Float64).alias(col))


# check the number of 0s and percentage after replacing with None 
for col in string_cols: # count the number of NAs
    na_count = df.filter(pl.col(col).is_null()).height
    print(f'"NA" count: {na_count} / {total_rows}')
    print(f"percentage of nulls for {col}: {na_count / total_rows * 100}")
    neg_count = df.filter(pl.col(col) < 0).height
    print(f"negatives: {neg_count}")
    print(f"zeroes: {df.filter(pl.col(col) == 0).height} / {total_rows}")
```

```python
# Check the columns with negative values 
string_cols = [col for col, dtype in df.schema.items() if dtype == pl.Utf8]
numeric_cols = [col for col in df.columns if col not in string_cols]

print(f"Number of numeric cols: {len(numeric_cols)}, Number of String cols: {len(string_cols)}")

# check for negatives columns  
negative_cols = [
    col for col in numeric_cols
    if df.select((pl.col(col) < 0).any()).item() 
]
df_negs = df.select(negative_cols)
print("Negative cols:")
print(df_negs.columns)

# check the number of negative values in each col 
for col in negative_cols:
    num_negatives = (df[col] < 0).sum()
    total_rows = len(df)
    percent_negative = num_negatives / total_rows * 100
    print(f"percent of negatives for {col}: {percent_negative}")
    print(f"Number of negatives: {num_negatives} / {total_rows}")

    # Filter to negative values
    negatives = df.select(pl.col(col).filter(pl.col(col) < 0))

    n_unique_neg = negatives.unique().height
    n_total_neg = negatives.height

    # Check if all negative values are unique
    if n_unique_neg == n_total_neg:
        print("All negative values are unique")
    else:
        print("There are duplicate negative values")

    zero_count = df.filter(pl.col(col) == 0).height
    print(f"Number of zeroes for {col}: {zero_count} / {total_rows}")
    print(f"Percent of zeroes for {col}: {zero_count / total_rows * 100}")
```

```python
# impute all columns with invalid values first 
limit = 1.0 
cols = ['HSWH040S', 'HSWH041S', 'ECYPTAMED', 
        'ECYPMAMED', 'ECYPFAMED', 'ECYHTAMED', 'ECYHMAMED', 'ECYHFAMED', 'ECYMTNMED'] # all cols that contain invalid values 

for col in cols:
    valid = df.select(pl.col(col).drop_nulls())
    mean_val = valid.select(pl.col(col).mean()).item()
    median_val = valid.select(pl.col(col).median()).item()

    # assume median for imputation if distribution is not symmetric 
    impute_value = mean_val if abs(mean_val - median_val) < limit else median_val

    df = df.with_columns(pl.col(col).fill_null(impute_value))
```

```python
# drop the columns used for computing the target variable for the unsupervised dataset
df = df.drop(drop_later)
df.width # check the width after dropping 
```

```python
# check for outliers for every column 
threshold = 3.0 # threshold for z score 
i = 1
for col in df.columns:
    mean = df.select(pl.col(col).mean()).item()
    std = df.select(pl.col(col).std()).item()
    median = df.select(pl.col(col).median()).item() 
    count = len(df)
    
    # Add a z score column and compute z score for each entry for that column 
    df = df.with_columns(((pl.col(col) - mean) / std).alias(f"{col}_zscore"))

    # filter out outliers 
    outliers = df.filter(pl.col(f"{col}_zscore").abs() > threshold)
    
    # count the number of outliers compared to the col 
    count_ratio = round(outliers.height / count * 100, 2)
    if count_ratio < 10.0 and count_ratio > 0.0: # if there are less than 10% of outliers 
        print(f"Imputing reasonable outliers for column {i} --- {col}: {count_ratio:.2f}%")
        # impute using mean or median 
        impute_val = mean if abs(mean - median) < 1.0 else median
        df = df.with_columns(
            pl.when(pl.col(f"{col}_zscore").abs() > threshold)
            .then(impute_val)
            .otherwise(pl.col(col))
            .alias(col)
        )
        
    elif count_ratio >= 10.0: # drop outliers otherwise if the % of outliers is larger than 10% 
        # drop rows where the z-score > threshold 
        print("Found {col} where number of outliers >= 10%, {count_ratio} %")
        df = df.filter(pl.col(f"{col}_zscore").abs() <= threshold)
    
    df = df.drop(f"{col}_zscore") # drop the column after finishing outliers 
    i += 1
```
